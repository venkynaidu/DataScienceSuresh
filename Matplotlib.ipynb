```python
import numpy as np
t=np.arange(0.,5.,0.2)
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.legend()
plt.grid()
plt.show()
```

```python
use the setp() :  set properties functions
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='b',linewidth=8.0)
plt.setp(lines[1],'color','g','linewidth',2.0)
plt.grid()
plt.show()
```

# Working with multiple figures and axes:

```python
def f(t): 
    return np.exp(-t)**np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(1)

plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b--')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
#modifying of values in figures
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(211)
plt.title("Hello this is title")
plt.figure(2)
plt.title("Hello this is title")
plt.show()
```
